```python
# importing packages

import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# reading image
img_org = cv2.imread("TajMahal.jpg") 
```

# SIFT

```python
# Defining a function to match keypoints between original image and transformed image

def apply_sift(original_image, modified_images):
    img_org = original_image.copy()
    sift = cv2.xfeatures2d.SIFT_create()
    img_org_gray = cv2.cvtColor(img_org, cv2.COLOR_BGR2GRAY)
    images_kp_match = {'imgs':[], 'img_match_score':[]}
    if len(modified_images)>1 and len(modified_images)<25:
      for img_mod in modified_images:

          img_mod_gray = cv2.cvtColor(img_mod, cv2.COLOR_BGR2GRAY)

          keypoints_org, descriptors_org = sift.detectAndCompute(img_org_gray, None)
          keypoints_mod, descriptors_mod = sift.detectAndCompute(img_mod_gray, None)

          # feature matching
          bf = cv2.BFMatcher(cv2.NORM_L1, crossCheck=True)

          matches = bf.match(descriptors_org, descriptors_mod)
          matches = sorted(matches, key = lambda x:x.distance)

          kp_match_img = cv2.drawMatches(img_org, keypoints_org, img_mod, keypoints_mod, matches[:50], img_mod, flags=2)
          img_kp_match_score = (len(matches)/len(keypoints_org))*100

          images_kp_match['imgs'].append(kp_match_img)
          images_kp_match['img_match_score'].append(round(img_kp_match_score, 2))
      
      avg_match_score = sum(images_kp_match['img_match_score'])/3
      images_kp_match['avg_match_score'] = round(avg_match_score, 2)
    else:
      img_mod_gray = cv2.cvtColor(modified_images, cv2.COLOR_BGR2GRAY)

      keypoints_org, descriptors_org = sift.detectAndCompute(img_org_gray, None)
      keypoints_mod, descriptors_mod = sift.detectAndCompute(img_mod_gray, None)

      # feature matching
      bf = cv2.BFMatcher(cv2.NORM_L1, crossCheck=True)

      matches = bf.match(descriptors_org, descriptors_mod)
      matches = sorted(matches, key = lambda x:x.distance)

      kp_match_img = cv2.drawMatches(img_org, keypoints_org, modified_images, keypoints_mod, matches[:50], modified_images, flags=2)
      img_kp_match_score = (len(matches)/len(keypoints_org))*100

      images_kp_match['imgs'] = kp_match_img
      images_kp_match['img_match_score'] = round(img_kp_match_score, 2)


    return images_kp_match
```

### A. Scaling

```python
def scale(image):
    img = image.copy()
    ## half scaling
    half_img = cv2.resize(img, (img.shape[0]//2, img.shape[1]//2))
    ## double scaling
    double_img = cv2.resize(img, (img.shape[0]*2, img.shape[0]*2))
    # stretching
    strech_img = cv2.resize(img, (1666, 2500), interpolation=cv2.INTER_LINEAR)
    return [half_img, double_img, strech_img]
```

```python
# Applying SIFT: Finding Keypoints between original image and every scaled images 

scaled_imgs = scale(img_org)
kp_match_images = apply_sift(img_org, scaled_imgs)

print('Average Keypoint match score for scaling trandormation: ', kp_match_images['avg_match_score'])
plt.figure(figsize=(15, 15))
for i in range(len(scaled_imgs)):
  img = kp_match_images['imgs'][i]
  match_score = kp_match_images['img_match_score'][i]
  plt.subplot(len(scaled_imgs), 1, i+1)
  plt.title(match_score)
  plt.xticks([]), plt.yticks([])
  plt.imshow(img)  
plt.show()
```

### B. Rotation

```python
def rotate(image):
  img = image.copy()
  ## grab the dimensions of the image and calculate the center of the image
  row, cols, _ = image.shape
  centerX, centerY = (cols//2, row//2)
  rotated_imgs = []
  ## rotate our image by various degrees around the center of the image
  for theta in np.arange(45, 360, 45):
    theta_transform_M = cv2.getRotationMatrix2D((centerX, centerY), theta, 1.0)
    theta_rotated_img = cv2.warpAffine(img, theta_transform_M, (row, cols))
    rotated_imgs.append(theta_rotated_img)

  return rotated_imgs
```

```python
# Applying SIFT: Finding Keypoints between original image and every rotated images 

rotated_imgs = rotate(img_org)
kp_match_images = apply_sift(img_org, rotated_imgs)

print('Average Keypoint match score for totation trandormation: ', kp_match_images['avg_match_score'])
plt.figure(figsize=(15, 15))
for i in range(len(rotated_imgs)):
  img = kp_match_images['imgs'][i]
  match_score = kp_match_images['img_match_score'][i]
  plt.subplot(len(rotated_imgs), 1, i+1)
  plt.title(match_score)
  plt.xticks([]), plt.yticks([])
  plt.imshow(img)  
plt.show()
```

```python
print('Average Keypoint match score for rotation trandormation: ', kp_match_images['avg_match_score'])
plt.figure(figsize=(50, 50))
for i in range(len(rotated_imgs)):
  img = kp_match_images['imgs'][i]
  match_score = kp_match_images['img_match_score'][i]
  plt.subplot(len(rotated_imgs), 1, i+1)
  plt.title(match_score)
  plt.xticks([]), plt.yticks([])
  plt.imshow(img)  
plt.show()
%matplotlib inline
```

### C. Affine

```python
def affine(image):
  img = image.copy()
  rows, cols, _ = img.shape
  # setting points on input and output image
  input_img_point = np.float32([[50, 50],
                                [200, 50],
                                [50, 200]])
  
  output_img_point = np.float32([[10, 100],
                                  [200, 50],
                                  [100, 250]])
              
  transform_matrix = cv2.getAffineTransform(input_img_point, output_img_point)
  affine_img = cv2.warpAffine(img, transform_matrix, (cols, rows))

  return affine_img
```

```python
affine_imgs = affine(img_org)
kp_match_images = apply_sift(img_org, affine_imgs)
```

```python
plt.figure(figsize=(20, 20))
img = kp_match_images['imgs']
match_score = kp_match_images['img_match_score']
plt.title(match_score)
plt.xticks([]), plt.yticks([])
plt.imshow(img)  
plt.show()
%matplotlib inline
```

### D. Perspective

```python
def perspective(image):
  img = image.copy()
  rows, cols, _ = img.shape
  pts1 = np.float32([[0, 260], [640, 260],
                      [0, 400], [640, 400]])
  pts2 = np.float32([[0, 0], [400, 0],
                      [0, 640], [400, 640]])
    
  # Apply Perspective Transform Algorithm
  persptv_matrix = cv2.getPerspectiveTransform(pts1, pts2)
  persptv_img = cv2.warpPerspective(img, persptv_matrix, (rows, cols))
  return persptv_img
```

```python
persptv_imgs = perspective(img_org)
kp_match_images = apply_sift(img_org, persptv_imgs)
```

```python
plt.figure(figsize=(20, 20))
img = kp_match_images['imgs']
match_score = kp_match_images['img_match_score']
plt.title(match_score)
plt.xticks([]), plt.yticks([])
plt.imshow(img)  
plt.show()
%matplotlib inline
```
